# Aesthetics

```r
# displacement and highway miles per gallon

#qplot() is depricated, it says to learn ggplot()
qplot(displ, hwy, data = mpg, color = drv)
ggplot(data = mpg, aes(displ, hwy, color = drv)) + geom_point()
```

```r
#qplot() is depricated, it says to learn ggplot()
qplot(displ, hwy, data = mpg, geom = c("point", "smooth"))
ggplot(data = mpg, aes(displ, hwy)) + 
    geom_point() +
    geom_smooth(method = "loess")
```

# Histograms

```r
#qplot() is depricated, it says to learn ggplot()

qplot(hwy, data = mpg, fill = drv, binwidth = 2)

ggplot() + geom_histogram(data = mpg, aes(hwy, fill = drv), binwidth = 2)
```

# Side-by-side Boxplots

```r
#qplot() is depricated, it says to learn ggplot()

qplot(drv, hwy, data = mpg, geom = "boxplot")

ggplot() + geom_boxplot(data = mpg, aes(drv, hwy))
```

# Facets

```r
#qplot() is depricated, it says to learn ggplot()

qplot(hwy, data = mpg, facets = drv ~ ., binwidth = 2)

ggplot() + 
    geom_histogram(data = mpg, aes(hwy), binwidth = 2) +
    facet_grid(drv ~ .)
```

```r
#qplot() is depricated, it says to learn ggplot()

qplot(displ, hwy, data = mpg, facets = . ~ drv) + geom_smooth()

ggplot(data = mpg, aes(displ, hwy)) + 
    geom_point() +
    geom_smooth(method = "loess") +
    facet_grid(. ~ drv)
```
