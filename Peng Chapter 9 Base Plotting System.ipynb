# Simple Base Graphics

```r
library(datasets)

str(airquality)
hist(airquality$Ozone)
```

```r
boxplot(airquality$Ozone ~ airquality$Month)

#OR
# boxplot(Ozone ~ Month, airquality)
```

```r
library(dplyr)

with(airquality, plot(Wind, Ozone))

#OR
# plot(airquality$Wind, airquality$Ozone)

title(main = 'Ozone and Wind in New York City')
with(filter(airquality, Month == 5), points(Wind, Ozone, col='red'))
```

```r
# type = 'n' let's you start with an empty plot, then you can draw the points in after
with(airquality, plot(Wind, Ozone, type = 'n'))
with(filter(airquality, Month == 5), points(Wind, Ozone, col = 'blue'))
with(filter(airquality, Month != 5), points(Wind, Ozone, col = 'red'))
title(main = "Ozone and Wind in New York City")
legend("topright", legend = c('May', 'Other Months'), pch = 1, col = c('blue', 'red'))
```

```r
with(airquality, plot(Wind, Ozone, pch=20, cex = 1.5))
# pch = 20 is filled in circles, cex = size of markers
title(main = 'Ozone and Wind in New York City')

model <- lm(Ozone ~ Wind, airquality) #linear regression, Ozone predicted by Wind
abline(model, lwd = 2) #abline can understand model object to be able to plot it
```

```r
options(repr.plot.width=14)

par(mfrow = c(1,2))
with(airquality, {
    plot(Wind, Ozone, main = "Ozone and Wind")
    plot(Solar.R, Ozone, main = "Ozone and Solar Radiation")
})
```

```r
options(repr.plot.width=14, repr.plot.height=5)

par(mfrow = c(1,3), oma=c(0,0,2,0))
with(airquality, {
    plot(Wind, Ozone, main = "Ozone and Wind")
    plot(Solar.R, Ozone, main = "Ozone and Solar Radiation")
    plot(Temp, Ozone, main = "Ozone and Temperature")
    mtext("Ozone and Weather in New York City", outer = TRUE)
})
```
